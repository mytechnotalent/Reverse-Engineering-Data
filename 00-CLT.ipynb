### Create a function to calc average.

```python
def my_avg(n):
    """
    Calculate the average of a list of numbers.

    Parameters:
        n (list): A list of numbers.

    Returns:
        float: The average of the numbers in the list.
    """
    return sum(n) / len(n)
```

## Central Limit Theorem


$$ s^2 = \frac{1}{N-1} \sum_{n=1}^{N} \left( \left(x_n - \bar{x}\right)^2 \right) $$

1. Calculate the sample average.
2. Calculate the deviation or difference of each measurement from the average.
3. Square the deviation.
4. Sum the squared deviation across all measurements.
5. Divide the sum of the squared deviation from the average.

### **The variance is the mean squared deviation from the average.**

### Let's create a variance function.

```python
def sample_variance(data):
    """
    Calculate the sample variance of a list of numbers (Bessel's correction: divide by n-1).

    Parameters:
        data (list): A list of numbers.

    Returns:
        float: The sample variance of the numbers.
    """
    mean = sum(data) / len(data)
    squared_deviation = [(x - mean) ** 2 for x in data]
    return sum(squared_deviation) / (len(data) - 1)
```

### Determine the variance of the average.

```python
sample_variance([5, 10, 15])
```

### **The standard deviation is the half width of the bell curve and is the square root of the variance.**

### Let's make a function to calculate the sample variance.

```python
def sample_sd(data):
    """
    Calculate the sample standard deviation of a list of numbers (using Bessel's correction, n-1).

    Parameters:
        data (list): A list of numbers.

    Returns:
        float: The sample standard deviation of the numbers.
    """
    return sample_variance(data) ** 0.5
```

```python
sample_sd([5, 10, 15]) 
```

### Visualize how the sample variance and sample standard deviation work.

```python
# Data
data = [5, 10, 15]
mean = sum(data) / len(data)
n = len(data)
x = data

# Calculate squared deviations and SD
sq_diffs = [(val - mean)**2 for val in data]
variance = sum(sq_diffs) / (n - 1)
sample_sd_ = variance ** 0.5

# Plot points and mean
plt.figure(figsize=(12, 6))
plt.scatter(x, 
            data, 
            color='blue', 
            s=120, 
            label='Data Points')
plt.axhline(mean, 
            color='red', 
            linestyle='--', 
            lw=2, 
            label='Mean = {:.1f}'.format(mean))

# Draw deviation lines and annotate
for i, y in enumerate(data):
    plt.plot([x[i], 
              x[i]], 
              [mean, y], 
              'k:', lw=1)
    diff = y - mean
    # Purple annotation
    plt.text(x[i], 
             y+1 if y>mean else y-2, 
             '({}-10)={:.1f}'.format(y, diff), 
             color='purple', 
             fontsize=8)
    # Orange annotation, just 0.5 units above/below the purple one
    plt.text(x[i], 
             (y+1.5 if y>mean else y-2.5), 
             '({}-10)²={:.1f}'.format(y, diff**2), 
             color='orange', 
             fontsize=8)

# SD calculation steps
plt.text(10, 
         4.5, 
         'n = {}'.format(n), 
         fontsize=13, 
         color='black', 
         ha='center')
plt.text(10, 
         3.3, 
         'Squared Deviations Sum = {} = {}'.format("+".join(str(int(sd)) for sd in sq_diffs), int(sum(sq_diffs))), 
         color='black', 
         fontsize=14, 
         ha='center')
plt.text(10, 
         2.0, 
         'Sample Variance = {}/(n-1) = {}/2 = {:.1f}'.format(int(sum(sq_diffs)), int(sum(sq_diffs)), variance), 
         color='black', 
         fontsize=14, 
         ha='center')
plt.text(10, 
         0.5, 
         'Sample SD = sqrt({:.1f}) = {:.2f}'.format(variance, sample_sd_), 
         color='darkgreen', 
         fontsize=15, 
         fontweight='bold', 
         ha='center')

# Calculation of Sample Standard Deviation
plt.title('Calculation of Sample Standard Deviation for [5, 10, 15]')
plt.xticks(data, 
           [str(v) for v in data])
plt.xlim(2, 18)
plt.ylim(0, 17)
plt.legend(loc='upper left')
plt.grid(False)
plt.tight_layout()
plt.show()
```

### Variability based on different sample sizes.

### Let's replicate 2500 times calculating the average from a sample size to ensure we have a normal distribution for the sample average.

```python
sample_sd([1, 2, 3])
```

```python
random.seed(42)

study_sd_vs_sample_size = [sample_sd([gen_samples_calc_avg(sample) for _ in range(2500)]) for sample in sample_size_study]
```

### Show the standard deviation of the average with respect to sample size.

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot( sample_size_study, study_sd_vs_sample_size, '-o')
ax.set_xscale('log')
ax.set_xlabel('sample size')
ax.set_ylabel('standard deviation of the average')
ax.grid(True)
plt.show()
```

### **The standard deviation on the average decreases as the sample size increases.**

### **The standard deviation on the average is called the `standard error on the mean` or `SEM`.**

### SEM = (standard_deviation) / (square root of the number of samples)

$$ \mathrm{SEM} = \frac{\sigma}{\sqrt{N}} $$

The $\sigma$ is the standard deviation of the measurements. We are currently using a random number generator that generates **uniform** random numbers. There is a known formula for the **variance** of a **uniform** distribution.

Let's assume the **lower** bound is $a$ and the **upper** bound is $b$. The formula for the **variance** of a **uniform distribution** is:

$$ variance = \frac{1}{12} \left( b - a \right)^2 $$

We have worked with a Uniform between 0 and 1, thus the Variance is:

$$ variance = \frac{1}{12} $$

The standard deviation is the square root of the variance:

$$ \sigma = \frac{1}{\sqrt{12}} $$

The SEM for our example is then:

$$ \mathrm{SEM} = \frac{1}{\sqrt{12 \times N} } $$

or

$$ \mathrm{SEM} = \left( \frac{ 1/12} {N} \right)^{0.5} $$

Let's use this formula and compare this theoretical to our simulation results...

```python
sem_theory = [((1/12) / sample) ** 0.5 for sample in sample_size_study]
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot( sample_size_study, study_sd_vs_sample_size, '-o', label='simulation')
ax.plot( sample_size_study, sem_theory, ':sr', label='theory')
ax.set_xscale('log')
ax.set_xlabel('sample size')
ax.set_ylabel('standard deviation of the average')
ax.grid(True)
ax.legend()
plt.show()
```

### ***WE CAN EXPRESS CONFIDENCE USING THE CLT AND THE STANDARD ERROR.***

### ***The confidence interval represents where you anticipate the true population average to be, based on your sample data and the chosen confidence level.***

### The most common confidence interval is 95%.

```python
sample_sd_ = sample_sd(small_sample)
```

```python
sample_sd_
```

```python
sem = sample_sd_ / (len(small_sample) ** 0.5)
```

```python
sem
```

```python
conf_int = (small_sample_avg - 2 * sem, small_sample_avg + 2 * sem)
```

```python
conf_int
```

```python
small_sample_avg
```

```python
conf_int[1] - conf_int[0]
```
